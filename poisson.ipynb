```python
import numpy as np
# import scipy.stats as stats
import pandas as pd # Import the pandas library

# Set parameter dan jumlah data
lambda_poisson = 8
n = 150  # Anda dapat menyesuaikan jumlah sampel sesuai kebutuhan

# Generate data dari distribusi Poisson
data = np.random.poisson(lambda_poisson, n)

print("Data Poisson:", data)
print("Jumlah Data yang Dibangkitkan:", len(data))

# Membuat DataFrame untuk menyimpan hasil
df = pd.DataFrame({'Jumlah Keberhasilan dalam 150 Percobaan': data})

# Menampilkan DataFrame
excel_file = 'poisson_distribution_data.xlsx'
# df = pd.read_excel(excel_file)
df.to_excel(excel_file, index=False)

print("Data berhasil disimpan dalam file Excel:", excel_file)
```

```python
import numpy as np
import scipy.stats as stats

poisson_data = df['Jumlah Keberhasilan dalam 150 Percobaan']

# Hitung statistik deskriptif
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)[0]
percentile_25 = np.percentile(data, 25)
percentile_75 = np.percentile(data, 75)
variance = np.var(data)
skewness = stats.skew(data)
kurtosis = stats.kurtosis(data)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Percentil ke-25: {percentile_25}")
print(f"Percentil ke-75: {percentile_75}")
print(f"Varians: {variance}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis}")
```

```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import matplotlib.pyplot as plt # Import the plotting library


# --- Your existing code ---
np.random.seed(0)

# Define or regenerate the 'data' variable before using it in the histogram:
lambda_poisson = 8  # Assuming lambda_poisson is still 8
n = 150  # Assuming n is still 150
data = np.random.poisson(lambda_poisson, n) # Regenerate data

# Now you can use these variables in the plotting code:
plt.hist(data, bins=30, alpha=0.7, color='blue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean}')
plt.axvline(median, color='orange', linestyle='dashed', linewidth=2, label=f'Median: {median}')
plt.axvline(mode, color='green', linestyle='dashed', linewidth=2, label=f'Mode: {mode}')
plt.axvline(percentile_25, color='purple', linestyle='dashed', linewidth=2, label=f'25th Percentile: {percentile_25}')
plt.axvline(percentile_75, color='brown', linestyle='dashed', linewidth=2, label=f'75th Percentile: {percentile_75}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram with Mean, Median, Mode, and Percentiles')
plt.legend()
plt.show()

# Print variance, skewness, and kurtosis
print(f"Variance: {variance:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurtosis:.2f}")
```

```python
import matplotlib.pyplot as plt

# Hitung frekuensi
unique, counts = np.unique(data, return_counts=True)
relative_frequencies = counts / n

# Tampilkan frekuensi relatif
for value, freq in zip(unique, relative_frequencies):
    print(f"Nilai: {value}, Frekuensi Relatif: {freq}")

# Buat grafik PMF
plt.bar(unique, relative_frequencies)
plt.xlabel('Jumlah Gelembung')
plt.ylabel('Frekuensi Relatif')
plt.title('Probability Mass Function (PMF)')
plt.show()
```

```python
# Hitung CDF
cdf = np.cumsum(counts) / n

# Plot CDF
plt.plot(unique, cdf, marker='o', linestyle='-', color='b')
plt.xlabel('Jumlah Gelembung')
plt.ylabel('Probabilitas Kumulatif')
plt.title('Cumulative Distribution Function (CDF)')
plt.grid(True)
plt.show()
```
